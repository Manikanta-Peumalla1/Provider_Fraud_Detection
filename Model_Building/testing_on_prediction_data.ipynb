```python
import pickle
import pandas as pd
import numpy as np
```

### Load saved model in the training phase

```python
with open("../Models/lgbm_desired.pkl","rb") as f:
    model=pickle.load(f)
```

```python
data=pd.read_csv("../Model_Input_Data/prediction_input_data.csv")
data.head()
```

```python
x_var=data.reindex(columns=model.feature_names_in_,fill_value=0)
```

```python
predictions=model.predict(x_var)
prob=model.predict_proba(x_var)
prob_fraud=[row[1] for row in prob]
provider=data['Provider']
results=pd.DataFrame({'Provider':provider,'predictions':predictions,'Fraud_Probability':prob_fraud})
results['Fraud_Probability']=results['Fraud_Probability'].apply(lambda x:round(x,3))
results
```
